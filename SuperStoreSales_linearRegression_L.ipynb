```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
pf['Ship Mode'] = le.fit_transform(pf['Ship Mode'])
pf['Country'] = le.fit_transform(pf['Country'])
pf['Region'] = le.fit_transform(pf['Region'])
pf['Segment'] = le.fit_transform(pf['Segment'])
pf['Sub-Category'] = le.fit_transform(pf['Sub-Category'])
pf['Category'] = le.fit_transform(pf['Category'])
pf['Product ID'] = le.fit_transform(pf['Product ID'])
pf['Payment Mode'] = le.fit_transform(pf['Payment Mode'])
pf
```

```python
x = pf.drop(['Profit'], axis = 1).values
y = pf['Profit'].values
x
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler
sr = StandardScaler()
sr.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y, train_size= 0.3, random_state=117)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
mae = mean_absolute_error(y_test,y_pred)
mae
```

```python
mse = mean_squared_error(y_test,y_pred)
mse
```

```python
r2_score(y_test,y_pred)
```
